```python
import jsonlines
phase=[]
table_id=[]
question=[]
sql=[]
with jsonlines.open("/content/drive/My Drive/Colab Notebooks/RoBERTa NL2SQL/train.jsonl") as f:
  for line in f.iter():
    phase.append(line['phase'])
    table_id.append(line['table_id'])
    question.append(line['question'])
    sql.append(line['sql'])
import pandas as pd
d={'phase':phase,'table_id':table_id,'question':question,'sql':sql}
train_data=pd.DataFrame(d)
train_data
f.close()


train_data.columns=['phase','id','question','sql']


import jsonlines
id=[]
header=[]
types=[]
rows=[]
with jsonlines.open("/content/drive/My Drive/Colab Notebooks/RoBERTa NL2SQL/train.tables.jsonl") as f:
  for line in f.iter():
    id.append((line['id']))
    header.append((line['header']))
    types.append((line['types']))
    rows.append(line['rows'])
d={'id':id,'header':header,'types':types,'rows':rows}
train_table_info=pd.DataFrame(d)
train_table_info

```

```python
merged_data=train_data.merge(train_table_info,on='id')
merged_data
```

```python
from tqdm import tqdm
sql_in_text=[]
aggregate=['','MAX','MIN','COUNT','SUM','AVG']
operator=['=','<','>']
word_operator=['eq','lt','gt']
op=[]
for i in tqdm(range(len(merged_data))):
  sql=merged_data.sql.iloc[i]
  col=merged_data.header.iloc[i][sql['sel']]
  agg=aggregate[sql['agg']]
  cond_col=[]
  cond_opr=[]
  cond_value=[]
  for j in sql['conds']:
    cond_col.append(merged_data.header.iloc[i][j[0]])
    cond_opr.append(word_operator[j[1]])
    cond_value.append(j[2])
  resulted_sql="select "
  if(agg==''):
    op.append("noop")
    resulted_sql+=col+" where "
  else:
    op.append(agg.lower())
    resulted_sql+=agg+"("+col+") where "

  for j in range(len(cond_col)):
    if(j!=0):
      resulted_sql+="and "
    resulted_sql+=cond_col[j]+' '+cond_opr[j]+' '
    resulted_sql+=str(cond_value[j])+' '

  sql_in_text.append(resulted_sql)




merged_data['operation']=op
merged_data['resulted_sql']=sql_in_text


merged_data

```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Load your training, validation, and test datasets (X_train, y_train, X_valid, y_valid, X_test, y_test)

# Step 2: Training
# Combine training and validation datasets for training the final model
X_train_combined =pd.concat([merged_data['question'],merged_paratr_df['paraquestion'],val_merged_data['question'],merged_valp_df['paraquestion']])
y_train_combined = pd.concat([merged_data['operation'],merged_paratr_df['operation'],val_merged_data['operation'],merged_valp_df['operation']])

# Create feature vectors using CountVectorizer
vectorizer = CountVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train_combined)

# Train the Naive Bayes classifier
classifier = MultinomialNB()
classifier.fit(X_train_vectorized, y_train_combined)


#X_train_vectorized = vectorizer.transform(merged_data['question'])
X_train_vectorized = vectorizer.transform(pd.concat([merged_data['question'],merged_paratr_df['paraquestion']]))
y_pred_train = classifier.predict(X_train_vectorized)
#training_accuracy = accuracy_score(merged_data['operation'], y_pred_train)
training_accuracy = accuracy_score(pd.concat([merged_data['operation'],merged_paratr_df['operation']]), y_pred_train)
print(f"Training Accuracy: {training_accuracy:.2f}")

# Step 3: Validation
X_valid_vectorized = vectorizer.transform(pd.concat([val_merged_data['question'],merged_valp_df['paraquestion']]))
y_pred_valid = classifier.predict(X_valid_vectorized)

# # Evaluate performance on validation data
valid_accuracy = accuracy_score(pd.concat([val_merged_data['operation'],merged_valp_df['operation']]), y_pred_valid)
print(f"Validation Accuracy: {valid_accuracy:.2f}")


X_test_vectorized = vectorizer.transform(test_merged_data['question'][0:])
y_pred_test = classifier.predict(X_test_vectorized)

# # Evaluate performance on test data
test_accuracy = accuracy_score(test_merged_data['operation'], y_pred_test)
print(f"Test Accuracy: {test_accuracy:.2f}")

# # Print classification report for more detailed performance metrics
print("Classification Report for Test Data:")
print(classification_report(test_merged_data['operation'], y_pred_test))
```
